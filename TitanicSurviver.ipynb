```python
df1=df
df1.head()
```

Converting sex into binary value 

```python
df1["Sex"]=df1["Sex"].replace({"male":1,"female":0})
df1.head()

```

```python
df1["Pclass"].value_counts()
```

```python
plt.pie(df1["Pclass"].value_counts(),labels=["class 3","class 1","class 2"],autopct="%0.2f%%")
```

```python
df1.drop("Cabin",axis=1,inplace=True)
df1.drop("Embarked",axis=1,inplace=True)
df1.head()
# df1.columns
```

```python
x=df1[['Pclass', 'Sex', 'Age', 'SibSp',
       'Parch', 'Fare']]

y=df1["Survived"]
x.isnull().sum()
```

Using Train test split model with train data as 70% and test data as 30%

```python

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score as a
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.37,random_state=23)

```

Decison tree

```python
from sklearn.tree import DecisionTreeClassifier
DT=DecisionTreeClassifier(criterion='entropy',max_depth=3)
DT.fit(x_train,y_train)
y_pred=DT.predict(x_test)
acc=a(y_test,y_pred)
print("Accuracy of Decision tree",acc*100)
```

Random forest

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100)
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
acc=a(y_test,y_pred)
print("Accuracy of Random forest",acc*100)

```
